```python
#1. Write a python program to find Minkowskii Distance between two points.
from math import *
from decimal import Decimal
def p_root(value, root):
    root_value = 1 / float(root)
    return round (Decimal(value) **
            Decimal(root_value), 3)
def minkowski_distance(x, y, p_value):
    return (p_root(sum(pow(abs(a-b), p_value)
            for a, b in zip(x, y)), p_value))
vector1 = [0, 2, 3, 4]
vector2 = [2, 4, 3, 7]
p = 3
print(minkowski_distance(vector1, vector2, p))
```

```python
#2.Write a Python NumPy program to compute the weighted average along the specified axis of a given flattened array.
import numpy as np
a = np.arange(9).reshape((3,3))
print("Original flattened array:")
print(a)
print("Weighted average along the specified axis of the above flattened array:")
print(np.average(a, axis=1, weights=[1./4, 2./4, 2./4]))
```
