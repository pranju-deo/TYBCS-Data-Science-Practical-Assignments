```python
#5.Write a Python program to add a column to the dataframe “BMI” which is calculated as : weight/height2
import pandas as pd
data = [
  { 'h': 1.5, 'w': 70 },
  { 'h': 1, 'w': 60 },
  { 'h': 2.0, 'w': 50 },
]
df = pd.DataFrame(data)
print(df)
print("------------------")
def BMI(data):
    return data['w'] / data['h']**2
df['bmi'] = df.apply(BMI, axis=1)
print(df)
```
